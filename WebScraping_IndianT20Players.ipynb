```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

url = 'https://stats.espncricinfo.com/ci/engine/records/averages/batting.html?class=3;current=2;id=6;type=team'
response = requests.get(url)
soup = bs(response.text, "html.parser")

table = soup.find("table")
rows = table.find_all("tr")
data = []
d =[]
for i in rows[0].find_all('th'):
    d.append(i.text.strip())
del rows[0]
for row in rows:
    cols = row.find_all("td")
    cols = [col.text.strip() for col in cols]
    data.append(cols)

df = pd.DataFrame(data,columns=d)
df.head()
```

The experience of player is calculated using values in Span Column

```python
#the span is first split into a list containing start year and end year(current year)
df1 = df['Span'].apply(lambda s: s.split('-'))
```

```python
#the experience column is added to dataframe we are working on
df['Experience'] = df1.apply(lambda s: int(s[1])-int(s[0]))
```

```python
df.head()
```

```python
#the span column is dropped so as to avoid extra data
df = df.drop('Span',axis=1)
```

```python
df.head()
```

```python
#the null values are eliminated
df = df.replace('-',0)
```

```python
df.head()
```

```python
#The players with experience more than 10 years

Exp_10 = df[df['Experience']>=10]
```

```python
Exp_10.reset_index(drop = True)
```

```python
#The players with experience more than 10 years and scored one century (100) minimum
Exp_10[Exp_10['100'] !='0'].reset_index(drop = True)
```

```python
df['SR']=df['SR'].astype(float)
```

```python
#Average Strike Rate considering all the players.
#The data has too much inconsistency therefore median is used insted of mean.
avg_str = df['SR'].median()
```

```python
# The players who has maintained strike rate greater than average
df[df['SR']>=avg_str].reset_index(drop=True)
```
